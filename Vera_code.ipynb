```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sn
```

```python
df=pd.read_csv('Dataset.csv')
df
```

```python
fig = plt.figure(figsize = (13, 7))
#Count the number of datapoints attributed with each weight category
counts = df["NObeyesdad"].value_counts()
plt.bar(counts.index, counts.values, color="green")
plt.xlabel("Weight Classification")
plt.ylabel("Number of Respondents")
plt.title("Number of Respondents per Weight Classification")
plt.savefig('count.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(6, 6))

# Define the color palette for each category
colors = {'Female': 'pink','Male': 'cyan'}

# Create the grouped bar chart
g = sn.catplot(x="NObeyesdad", hue="Gender", col="FAVC", kind="count", palette=colors, data=df)
g.set_xticklabels(rotation=30)
plt.savefig('barfav.png', dpi=300, bbox_inches='tight')
#plt.title("Number of Respondents per Weight Classification for people with high caloric food consumption")

plt.show()
```
